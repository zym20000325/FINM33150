```python
def ratio_row(ticker,start_date,end_date,df_debt_to_market_cap,df_return_on_investment,df_price_to_earnings):

    df_zacks = calculate_ratio(ticker,start_date,end_date)

    print(len(df_zacks))
    print(len(df_debt_to_market_cap))

    df_debt_to_market_cap = df_debt_to_market_cap.drop(ticker, axis=1)
    df_debt_to_market_cap = df_debt_to_market_cap.join(df_zacks['debt_to_market_cap'])

    df_return_on_investment = df_return_on_investment.drop(ticker, axis=1)
    df_return_on_investment = df_return_on_investment.join(df_zacks['return_on_investment'])

    df_price_to_earnings = df_price_to_earnings.drop(ticker, axis=1)
    df_price_to_earnings = df_price_to_earnings.join(df_zacks['price_to_earnings'])

    # df_debt_to_market_cap[ticker] = df_zacks['debt_to_market_cap']
    # df_return_on_investment[ticker] = df_zacks['return_on_investment']
    # df_price_to_earnings[ticker] = df_zacks['price_to_earnings']

    return df_debt_to_market_cap,df_return_on_investment,df_price_to_earnings
```

```python
def ratio_dataframe(ticker_list,date_list,start_date,end_date):

    df_debt_to_market_cap = pd.DataFrame(0.0, index=date_list, columns=ticker_list)
    df_return_on_investment = pd.DataFrame(0.0, index=date_list, columns=ticker_list)
    df_price_to_earnings = pd.DataFrame(0.0, index=date_list, columns=ticker_list)

    count = 0

    for ticker in ticker_list[150:]:

        print(ticker)

        print(df_debt_to_market_cap.tail())

        df_debt_to_market_cap,df_return_on_investment,df_price_to_earnings = ratio_row(ticker,start_date,end_date,df_debt_to_market_cap,df_return_on_investment,df_price_to_earnings)

        print(df_debt_to_market_cap.tail())

        count += 1
        # if count%10 == 0:
        #     print(count)

    return df_debt_to_market_cap,df_return_on_investment,df_price_to_earnings
```
